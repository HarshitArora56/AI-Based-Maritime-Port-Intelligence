#  PHASE 8 – MACHINE LEARNING MODELING

```python
import os
import pandas as pd

BASE_PATH = "C:/Users/YASH/Desktop/maritime_project/AI-Based-Maritime-Port-Intelligence"
GOLD_PATH = os.path.join(BASE_PATH, "data", "gold")

df = pd.read_csv(os.path.join(GOLD_PATH, "ml_features_binary.csv"))
df.head()
```

```python
#Define Target & Features (NO LEAKAGE VERSION)
```

```python
TARGET = "is_congested"

X = df.drop(columns=[TARGET])
y = df[TARGET]
```

```python
#Train–Test Split (Stratified)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

rf_pipeline_reduced = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", RandomForestClassifier(
        n_estimators=200,          # fewer trees
        max_depth=2,               # VERY IMPORTANT
        min_samples_leaf=200,      # VERY IMPORTANT
        min_samples_split=500,
        max_features="sqrt",       # reduces correlation
        class_weight="balanced",
        random_state=45,
        n_jobs=-1
    ))
])

rf_pipeline_reduced.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

train_acc_rf = accuracy_score(
    y_train,
    rf_pipeline_reduced.predict(X_train)
)

test_acc_rf = accuracy_score(
    y_test,
    rf_pipeline_reduced.predict(X_test)
)

print("Random Forest (Regularized)")
print("Train Accuracy:", train_acc_rf)
print("Test Accuracy :", test_acc_rf)
```

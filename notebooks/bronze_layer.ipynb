```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Maritime-Bronze-Layer") \
    .master("local[*]") \
    .config("spark.sql.shuffle.partitions", "8") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")
```

```python
BASE_PATH = "E:\C-DAC\Major Project\maritime_project"

RAW_PATH = f"{BASE_PATH}/data/raw"
BRONZE_PATH = f"{BASE_PATH}/data/bronze"
```

```python
df1 = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(f"{RAW_PATH}/df1_ais_kattegat.csv")

df1.count()
df1.printSchema()
```

```python
df2 = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(f"{RAW_PATH}/df2_ais_global.csv")

df2.count()
df2.printSchema()
```

```python
df3 = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(f"{RAW_PATH}/df3_ais_unacorn.csv")

df3.count()
df3.printSchema()
```

```python
df4 = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv(f"{RAW_PATH}/df4_container_tracking.csv")

df4.count()
df4.printSchema()
```

```python
df1.write.mode("overwrite").parquet(f"{BRONZE_PATH}/ais_kattegat")
df2.write.mode("overwrite").parquet(f"{BRONZE_PATH}/ais_global")
df3.write.mode("overwrite").parquet(f"{BRONZE_PATH}/ais_unacorn")
df4.write.mode("overwrite").parquet(f"{BRONZE_PATH}/container_tracking")
```
